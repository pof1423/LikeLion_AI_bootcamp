```python
data = tada_eta[['ATA', 'api_eta', 'hour']]
data['distance'] = ((tada_eta['pickup_lat']-tada_eta['driver_lat'])**2 + (tada_eta['pickup_lng']-tada_eta['driver_lng'])**2)*10000
data.head()
```

### 2. Ordinal encoding vs. One-hot encoding

Ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder(dtype = np.int32)
ordinal = enc.fit_transform(np.asarray(tada_eta['pickup_gu']).reshape(-1,1))
```

```python
data['pickup_gu'] = ordinal[:, 0]
```

```python
data.head()
```

```python
train = data[:12000]
test = data[12000:]

x_tn = np.asarray(train.drop('ATA', 1))
x_te = np.asarray(test.drop('ATA', 1))
y_tn = np.asarray(train['ATA'])
y_te = np.asarray(test['ATA'])
```

One-hot encoding

```python
data2 = tada_eta[['ATA', 'api_eta', 'hour']]
data2['distance'] = ((tada_eta['pickup_lat']-tada_eta['driver_lat'])**2 + (tada_eta['pickup_lng']-tada_eta['driver_lng'])**2)*10000
data2.head()
pickup_gu = pd.get_dummies(tada_eta['pickup_gu'])

data2 = pd.concat([data2, pickup_gu], axis = 1)
data2.head()
```

```python
train2 = data2[:12000]
test2 = data2[12000:]

x_tn2 = np.asarray(train2.drop('ATA', 1))
x_te2 = np.asarray(test2.drop('ATA', 1))
y_tn2 = np.asarray(train2['ATA'])
y_te2 = np.asarray(test2['ATA'])
```

### 4. Regression tree with Grid search

Ordinal encoding

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'max_iter' : [50, 100, 150],
    'max_depth' : [2, 3, 4],
    'min_samples_leaf' : [5, 10],
    'learning_rate' : [0.05, 0.1],
    'categorical_features' : [3,]
}
```

```python
grid_hgbr = GridSearchCV(estimator=HistGradientBoostingRegressor(), param_grid = params,
                         cv = 3, n_jobs = -1,
                         scoring = 'neg_mean_squared_error', verbose = 2)
grid_hgbr.fit(x_tn, y_tn)
params_fit = grid_hgbr.best_params_
print(params_fit)
```

```python
model = HistGradientBoostingRegressor(**params_fit).fit(x_tn, y_tn)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print("Errors of Regression model with Ordinal encoding")
mse = mean_squared_error(y_te, model.predict(x_te))
print("The mean squared error(MSE) on test set: {:.4f}".format(mse))
print("The initial error of API ETA on test set: {:.4f}".format(mean_squared_error(y_te, x_te[:,0]) ))

mae = mean_absolute_error(y_te, model.predict(x_te))
print("The mean absolute error (MAE) on test set: {:.4f}".format(mae))
print("The initial error of API ETA on test set: {:.4f}".format(mean_absolute_error(y_te, x_te[:,0]) ))
```

One-hot encoding

```python
params2 = {
    'max_iter' : [50, 100, 150],
    'max_depth' : [2, 3, 4],
    'min_samples_leaf' : [2, 5, 10],
    'learning_rate' : [0.01, 0.05, 0.1]
}
```

```python
grid_hgbr2 = GridSearchCV(estimator=HistGradientBoostingRegressor(), param_grid = params2,
                         cv = 3, n_jobs = -1,
                         scoring = 'neg_mean_squared_error', verbose = 2)
grid_hgbr2.fit(x_tn2, y_tn2)
params_fit2 = grid_hgbr2.best_params_
print(params_fit2)
```

```python
model2 = HistGradientBoostingRegressor(**params_fit2).fit(x_tn2, y_tn2)
```

```python
print("Errors of Regression model with One-hot encoding")
mse2 = mean_squared_error(y_te2, model2.predict(x_te2))
print("The mean squared error(MSE) on test set: {:.4f}".format(mse))
print("The initial error of API ETA on test set: {:.4f}".format(mean_squared_error(y_te2, x_te2[:,0]) ))

mae2 = mean_absolute_error(y_te2, model2.predict(x_te2))
print("The mean absolute error (MAE) on test set: {:.4f}".format(mae))
print("The initial error of API ETA on test set: {:.4f}".format(mean_absolute_error(y_te2, x_te2[:,0]) ))
```
